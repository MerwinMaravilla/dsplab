```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 100 samples of the input signal x[n]
n = np.arange(100)
x_n = np.sin(0.2 * np.pi * n)

# Initialize the output signal y[n]
y_n = np.zeros(100)

# Difference equation: y[n] = 0.7 * y[n-1] + x[n]
for i in range(1, 100):
    y_n[i] = 0.7 * y_n[i-1] + x_n[i]

# Plot both input signal x[n] and system output y[n] together
plt.figure(figsize=(10, 6))

plt.plot(n, x_n, label='Input Signal x[n]', color='b')
plt.plot(n, y_n, label='System Output y[n]', color='r')
plt.title('Input Signal and System Output')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Show the combined plot
plt.tight_layout()
plt.show()

```

```python

n = np.arange(100)
x_n = np.sin(0.2 * np.pi * n)

# Step 2: Simulate the output of the non-linear system y[n] = x[n]^2
y_n = x_n ** 2

# Step 3: Scatter plot for the input signal
plt.figure(figsize=(10, 4))
plt.scatter(n, x_n, color='blue', label='Input Signal x[n] = sin(0.2πn)', alpha=0.6)
plt.title('Input Signal')
plt.xlabel('Sample Number (n)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()

# Step 4: Bar plot for the output signal
plt.figure(figsize=(10, 4))
plt.bar(n, y_n, color='orange', label='Output Signal y[n] = (x[n])^2', alpha=0.6)
plt.title('Output Signal')
plt.xlabel('Sample Number (n)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()

# Step 5: Normal plot for both signals in a single plot
plt.figure(figsize=(10, 4))
plt.plot(n, x_n, label='Input Signal x[n]', color='blue', linewidth=2)
plt.plot(n, y_n, label='Output Signal y[n]', color='orange', linewidth=2)
plt.title('Input and Output Signals')
plt.xlabel('Sample Number (n)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Define the parameters
np.random.seed(0)  # For reproducibility
n = np.arange(100)

# Generate clean input signal
Xclean_n = np.sin(0.3 * np.pi * n)

# Generate white Gaussian noise
noise = np.random.normal(0, 0.5, 100)

# Create the noisy input signal
Xnoisy_n = Xclean_n + noise

# Initialize the output signal y[n]
y_n = np.zeros(100)

# Difference equation: y[n] = 0.5 * y[n-1] + 0.5 * x[n]
for i in range(1, 100):
    y_n[i] = 0.5 * y_n[i-1] + 0.5 * Xnoisy_n[i]

# Plot the clean input signal, noisy input signal, and system output together
plt.figure(figsize=(10, 6))

plt.plot(n, Xclean_n, label='Clean Input Signal Xclean[n]', color='g')
plt.plot(n, Xnoisy_n, label='Noisy Input Signal Xnoisy[n]', color='b')
plt.plot(n, y_n, label='System Output y[n]', color='r')
plt.title('Clean Input Signal, Noisy Input Signal, and System Output')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Show the combined plot
plt.tight_layout()
plt.show()
```
